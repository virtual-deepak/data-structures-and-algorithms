```python
def can_attend_all_meetings(time_intervals):
    time_intervals = sorted(time_intervals, key=lambda time_interval: time_interval[0])
    print(f"Sorted time intervals: {time_intervals}")
    for index, _ in enumerate(time_intervals):
        if (index < (len(time_intervals) - 1)):
            if (time_intervals[index][1] > time_intervals[index + 1][0]):
                return False
    return True
```

```python
can_attend_all_meetings([[0, 30], [5, 10], [15, 20]])
```

```python
can_attend_all_meetings([[7, 10], [2, 4]])
```

```python
can_attend_all_meetings([[1, 2], [3, 4], [10, 11], [5, 6], [7, 8], [6, 10]])
```

**Time Complexity** = O(nlogn) as it includes sorting<br><br>
**Space Complexity** (aux) = O(1)
