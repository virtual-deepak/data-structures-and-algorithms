```python
from typing import List


def interesection_3_sorted_arrays(arr1: List[int], arr2: List[int], arr3: List[int]):
    result = []
    arr1_pointer = 0
    arr2_pointer = 0
    arr3_pointer = 0
    while arr1_pointer < len(arr1) and arr2_pointer < len(arr2) and arr3_pointer < len(arr3): 
        if (arr1[arr1_pointer] == arr2[arr2_pointer] == arr3[arr3_pointer]):
            result.append(arr1[arr1_pointer])
            arr1_pointer+=1
            arr2_pointer+=1
            arr3_pointer+=1
        elif ((arr1[arr1_pointer] <= arr2[arr2_pointer]) and (arr1[arr1_pointer] <= arr3[arr3_pointer])):
            arr1_pointer+=1
        elif ((arr2[arr2_pointer] <= arr1[arr1_pointer]) and (arr2[arr2_pointer] <= arr3[arr3_pointer])):
            arr2_pointer+=1
        elif ((arr3[arr3_pointer] <= arr1[arr1_pointer]) and (arr3[arr3_pointer] <= arr2[arr2_pointer])):
            arr3_pointer+=1
    return result
      
```

```python
interesection_3_sorted_arrays([1, 2, 3, 4, 5], [1, 2, 5, 7, 9], [1, 3, 4, 5, 8])
```

```python
interesection_3_sorted_arrays([1], [2], [3])
```

```python
interesection_3_sorted_arrays([1], [1], [1])
```

**Time Complexity** = O(n) where 'n' is the min length of the 3 arrays<br><br>
**Space Complexity** (aux) = O(n) for output array
